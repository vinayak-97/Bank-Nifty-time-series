```python
time_list = list(df["Time"])

min_15_time = []

first_time = time_list[0]

for i in range(1,len(time_list)):
    t1 = dt.datetime.strptime(str(first_time), '%H:%M:%S')
    t2 = dt.datetime.strptime('00:15:00', '%H:%M:%S')
    time_zero = dt.datetime.strptime('00:00:00', '%H:%M:%S')
    min_15_time.append((t1-time_zero+ t2).time())
    first_time = (t1-time_zero+ t2).time()
    
```

```python
# calculation of VWAP

df['TP'] = (df['High 5']+df['Low 5']+df['Close 5'])/3.0
```

```python
df['TradedValue']  = df['TP']*df['Volume 5']
```

```python
df['CumVolume'] = df['Volume 5'].cumsum()
```

```python
df['CumTradedValue'] = df['TradedValue'].cumsum()
```

```python
df['VWAP'] = df['CumTradedValue'] /df['CumVolume']
```
